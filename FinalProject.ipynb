```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
moviedb = pd.read_csv('/content/TMBD Movie Dataset.csv')
moviedb
```

```python
#Drop null or zero values
moviedb[['budget','revenue']] = moviedb[['budget','revenue']].replace(0,np.NAN)
moviedb.dropna(subset=['budget', 'revenue'], inplace=True)
```

```python
#del unwanted columns
del_col = ['imdb_id', 'homepage','tagline', 'keywords', 'overview', 'budget_adj','revenue_adj']
moviedb.drop(del_col, axis=1, inplace=True)
```

```python
from datetime import datetime
# Filter for movies from the last 15 years
current_year = datetime.now().year
past_15_years = current_year - 15
filtered_movies = moviedb[moviedb['release_year'] >= past_15_years]

# Calculate profitability (assuming columns for budget and box office revenue)
filtered_movies['profit'] = filtered_movies['revenue'] - filtered_movies['budget']

# Get the top 10 most profitable movies
top_profitable_movies = filtered_movies.nlargest(10, 'profit')

# Create the bar plot
sns.barplot(data=top_profitable_movies, x='original_title', y='profit', ci=None)
plt.xlabel('Movie Title')
plt.ylabel('Profit')
plt.title('Top 10 Most Profitable Movies in the Last 15 Years')
plt.xticks(rotation=90)
plt.show()
```

```python
# Ensure 'release_date' is in datetime format
moviedb['release_date'] = pd.to_datetime(moviedb['release_date'])

# Extract the month from the release date
moviedb['Month'] = moviedb['release_date'].dt.month_name()

# Count the number of releases per month
monthly_release_counts = moviedb['Month'].value_counts().reset_index()
monthly_release_counts.columns = ['Month', 'Release Count']

# Sort the months in order
monthly_release_counts['Month'] = pd.Categorical(monthly_release_counts['Month'])
monthly_release_counts = monthly_release_counts.sort_values('Month')

# Create the bar plot
sns.barplot(data=monthly_release_counts, x='Month', y='Release Count', ci=None)
plt.xlabel('Month')
plt.ylabel('Number of Movie Releases')
plt.title('Number of Movie Releases by Month')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by original_title and runtime, then get the average popularity
runtime_popularity = moviedb.groupby(['original_title', 'runtime'])['popularity'].mean().reset_index()

# Create a scatter plot
plt.figure(figsize=(12, 6))
sns.scatterplot(data=runtime_popularity, x='runtime', y='popularity', hue='original_title', legend=False)
plt.xlabel('Runtime (minutes)')
plt.ylabel('Average Popularity')
plt.title('Effect of Movie Runtime on Popularity')
plt.grid(True)
plt.show()
```

```python
#How does popularity affect revenue and profit
moviedb.groupby(['revenue','profit'])['popularity'].value_counts().sort_values(ascending=False)
```

```python
# Group by revenue and profit and calculate average popularity
popularity_by_revenue_profit = moviedb.groupby(['revenue', 'profit'])['popularity'].mean().reset_index()

# Create a scatter plot
plt.figure(figsize=(12, 6))
scatter = plt.scatter(data=popularity_by_revenue_profit, x='revenue', y='profit',
                      s=popularity_by_revenue_profit['popularity'] * 100,  # Scale point size by popularity
                      alpha=0.6, edgecolors='w', cmap='viridis')

# Add color bar
plt.colorbar(scatter, label='Average Popularity')
plt.xlabel('Revenue')
plt.ylabel('Profit')
plt.title('How Popularity Affects Revenue and Profit')
plt.xscale('log')  # Log scale for better visibility
plt.yscale('log')  # Log scale for better visibility
plt.grid(True)
plt.show()
```
